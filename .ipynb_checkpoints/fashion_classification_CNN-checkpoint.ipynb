<h2>Getting data from fashion mnist dataset</h2>

<p>
    The images are from 
    <a href='https://www.kaggle.com/zalando-research/fashionmnist'>here</a>
</p>
<p>
    The images are 28 by 28 pixels and only have one channel (meaning grayscale)
</p>
<p>
    In original train dataset, there are 60000 images and in original test dataset, there are 10000 images.
</p>

```python
import keras
from keras.datasets import fashion_mnist
import numpy as np
import matplotlib.pyplot as plt
from keras import preprocessing
```

```python
# getting train dataset and test dataset from fashion mnist
(X_train_orig, Y_train_orig), (X_test_orig, Y_test_orig) = fashion_mnist.load_data()
```

```python
# this is class label list and the index of each element corresponds to the number in Y_train_orig
# eg. T_shirt/top has 0 as its index so 0 in Y_train_orig corresponds to T_shirt/top
class_names_lst = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
X_train_orig.shape
```

```python
# Y_train_orig contains class label for each images.
# eg. Y_train_orig[0] will give 9, meaning Ankle Boot for X_train_orig[0]
Y_train_orig
```

```python
# shape of test set
X_test_orig.shape
```

```python
# just plotting random image from the train dataset
from random import randint

plt.imshow(X_train_orig[randint(0, X_train_orig.shape[0])], cmap='Greys')
```

```python
# this is just to plot to see how the images would look like for each category.
# here 0 means the class associated with it. So 0 refers to T_shirt/Top,...
# we can see from class_names_lst what those numbers refer to.

num_set = set()
count = 0
plt.figure(figsize=(10,10))
for i in range(30):
    if Y_train_orig[i] not in num_set:
        plt.subplot(5,5,count+1)
        plt.xticks([])
        plt.yticks([])
        plt.grid(False)
        
        plt.imshow(X_train_orig[i], cmap=plt.get_cmap('gray'))
        plt.xlabel(Y_train_orig[i])
        
        num_set.add(Y_train_orig[i])
        count += 1
    
plt.show()
```

```python
# reshape from (60000, 28, 28) to (60000, 28, 28, 1)
X_train = X_train_orig.reshape((X_train_orig.shape[0], 28, 28, 1))
X_test = X_test_orig.reshape((X_test_orig.shape[0], 28, 28, 1))

X_train.shape
```

```python
# changing dtype from uint8 to float32
Y_train = Y_train_orig.astype(np.float32)

Y_test = Y_test_orig.astype(np.float32)
```

<h2>Taking some images from test set</h2>

<h3>From 10,000 test set, I will be taking out randomly 100 images (10 for each set) for trying out on the web application</h3>

```python
X_test.shape
```

```python
import random

# just for testing
print(random.randrange(3, 9))
```

```python
# Here, we take random 100 images from the test set with each class having 10 images.

X_test_copy = np.copy(X_test)
Y_test_copy = np.copy(Y_test)

count_img_class = {}

test_images_data = {}

for i in range(10):
    count_img_class[i] = 0
    test_images_data[i] = []


class_count_indx = 0

while class_count_indx < len(class_names_lst):
    
    rand_img_indx = random.randrange(0, len(X_test_copy))
    
        
    while int(Y_test_copy[rand_img_indx]) != class_count_indx:
        rand_img_indx = random.randrange(0, len(X_test_copy))
    
    
    while count_img_class[int(Y_test_copy[rand_img_indx])] < 10:
      
        count_img_class[int(Y_test_copy[rand_img_indx])] += 1
        test_images_data[int(Y_test_copy[rand_img_indx])].append(X_test_copy[rand_img_indx])
        
        # remove image data from X_test_copy and Y_test_copy
        X_test_copy = np.delete(X_test_copy, rand_img_indx, 0) 
        Y_test_copy = np.delete(Y_test_copy, rand_img_indx, 0)
        
        rand_img_indx = random.randrange(0, len(X_test_copy))
        while int(Y_test_copy[rand_img_indx]) != class_count_indx:
            rand_img_indx = random.randrange(0, len(X_test_copy))
            
            
    
    class_count_indx += 1


    

print("X_test_copy shape: ", X_test_copy.shape)
print("Y_test_copy shape: ", Y_test_copy.shape)
print("\n")

for key in test_images_data:
    print("class {}: num of images: {}".format(key, len(test_images_data[key])))
```

```python
# display images that will be used on web
# these images are not saved yet

print("Images that will be used on web")

for key in test_images_data:
    for i in range(len(test_images_data[key])):
        image = test_images_data[key][i].reshape((28,28))
        
        plt.subplot(5,5,i+1)
        plt.xticks([])
        plt.yticks([])
        plt.grid(False)
        
        plt.imshow(image, cmap=plt.get_cmap('gray'))
        plt.suptitle("Class {}: {}".format(key, class_names_lst[key]))
        
    plt.show()
```

```python
# for simplicity, I try to put all corresponding images into corresponding folders.
# eg. I put all T_shirts/tops images into t_shirt folder
# these images will be used on the web.

dest_fold_names = {
    0: 't_shirt',
    1: 'trouser',
    2: 'pullover',
    3: 'dress',
    4: 'coat',
    5: 'sandal',
    6: 'shirt',
    7: 'sneaker',
    8: 'bag',
    9: 'ankle_boot'
}
```

```python
# I save those images for the web into respective folder with simple names
# eg. for first T_shirt/top, I would save in the folder t_shirt with filename t_shirt_1.png


file_extension = '.png'

new_test_img_path = 'new_test_img/'
for key in test_images_data:
    folder_path = dest_fold_names[key] + "/"
    for i in range(len(test_images_data[key])):
        file_name = dest_fold_names[key] + "_" + str(i+1) 
        
        complete_path = new_test_img_path + folder_path + file_name + file_extension
        
        image_to_be_saved = test_images_data[key][i].reshape((28,28,1))
        
        # save img
        preprocessing.image.save_img(complete_path, image_to_be_saved)
        
        
        # show images
        plt.subplot(5,5,i+1)
        plt.xticks([])
        plt.yticks([])
        plt.grid(False)
        
        plt.imshow(image_to_be_saved.reshape((28,28)), cmap=plt.get_cmap('gray'))
        plt.suptitle("Class {}: {}".format(key, class_names_lst[key]))
    plt.show()
```

```python
# copy the X_test_copy to X_test and Y_test_copy to Y_test
# I could have used X_test and Y_test directly without the use of X_test_copy or Y_test_copy
# After I fully tested that X_test_copy and Y_test_copy have number of images that I intend to have, I just copy those to X_test and Y_test.
# So X_test and Y_test will have 100 less than the original test set.

X_test = np.copy(X_test_copy)
Y_test = np.copy(Y_test_copy)

print("X_test shape: ", X_test.shape)
print("Y_test shape: ", Y_test.shape)
```

```python
from sklearn.metrics import confusion_matrix
import itertools

from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPool2D
from keras.optimizers import RMSprop,Adam
from keras.preprocessing.image import ImageDataGenerator
from keras.callbacks import ReduceLROnPlateau
```

```python
# in this dataset, dimension is 28 by 28
d = X_train.shape[1]

d
```

<h2>Experiment 1</h2>

```python
model = Sequential()


# first convolution layer
model.add(Conv2D(filters = 32, kernel_size = (3,3),padding='same', 
                 activation ='relu', input_shape = (d,d,1)))

model.add(MaxPool2D(pool_size=(2,2)))
model.add(Dropout(0.2))

model.add(Conv2D(filters = 64, kernel_size = (3,3),padding='same', 
                 activation ='relu'))

model.add(MaxPool2D(pool_size=(2,2)))
model.add(Dropout(0.2))



# fully connected
model.add(Flatten())

model.add(Dense(32, activation = "relu"))

model.add(Dense(10, activation = "softmax"))



# Define the optimizer
# lr = learning rate
optimizer = Adam(lr=0.001, beta_1=0.9, beta_2=0.999)

# Compile the model
model.compile(optimizer = optimizer , loss = "sparse_categorical_crossentropy", metrics=["accuracy"])



epochs = 20 
batch_size = 32

# Fit the model
history = model.fit(X_train, Y_train, epochs = epochs, validation_data = (X_test,Y_test), batch_size=batch_size)
```

```python
plt.plot(history.history['val_loss'], color='b', label="validation loss")
plt.plot(history.history['loss'], color='g', label="training loss")
plt.title("Loss")
plt.xlabel("Number of Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()

plt.plot(history.history['val_accuracy'], color='b', label="validation accuracy")
plt.plot(history.history['accuracy'], color='g', label="training accuracy")
plt.title("Accuracy")
plt.xlabel("Number of Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

<h2>Testing new test data</h2>

<p>
    I am testing with the test data reserved for use on the web because I want to see how the model does compared to the web model. In other words, I am just testing whether I have the right configuration on the web.
</p>
<p>
    At first, on the web, I try to classify 100 images taken from the test set and see how it does and compare to the one below. In this way, I can see clearly whether the model is doing what it is supposed to do.

<p>
    After I see that this model and the model on the web produces same accuracy, I know that I am doing correctly on the web and I take out random 30 images out of 100 images to be displayed to the users.
</p>

```python
# getting images that I have saved before and trying to see how many are correct out of 100 images

import os

data_dir = "./new_test_img/"
cwd = os.getcwd()
os.chdir(cwd)

folders_lst = os.listdir(data_dir)
print(folders_lst)


folder_name_dic = {'ankle_boot': 9, 'dress': 3, 'sneaker': 7, 'coat': 4, 'sandal': 5, 't_shirt': 0, 'bag': 8, 'shirt': 6, 'pullover': 2, 'trouser': 1}

incorrect_lst = []

correct_count = 0
for folder in folders_lst:
    if folder in folder_name_dic:
        each_folder_path = data_dir + folder + "/" 
        for file_name in os.listdir(each_folder_path):
            if '.png' in file_name:
                #print(file_name)
                
                complete_path = os.path.join(each_folder_path, file_name)
                
                new_img = preprocessing.image.load_img(complete_path, color_mode="grayscale", target_size=(d, d, 1))
                
                
                new_img_lst = preprocessing.image.img_to_array(new_img)
                
                rand_X_test_reshaped = new_img_lst.reshape((1, 28, 28, 1))
                
                #rand_X_test_reshaped = rand_X_test_reshaped
              
                # using model to predict class label
                Y_new_img = model_5.predict(rand_X_test_reshaped) 


                Y_new_img_class = np.argmax(Y_new_img,axis = 1) 

                if Y_new_img_class[0] == folder_name_dic[folder]:
                    correct_count += 1
                else:
                    incorrect_lst.append(file_name)
                
print("correct: ", correct_count)
print(incorrect_lst)
```
